# Calculate distributions of the kernel for the liquid and the solid

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
###################################################################
# Plot options
###################################################################

font = {'family' : 'serif',
        'serif'   : 'palatino',
        #'sans-serif'    : 'Computer Modern Sans serif', 
        'style'   : 'normal',
        'variant'   : 'normal',
        'stretch'   : 'normal',
        'weight'   : 'normal',
        'size'   : 20}
plt.rc('font', **font)
plt.rc('text', usetex=True)
plt.rcParams['figure.figsize'] = (15, 6)
```

Load data for the solid and the liquid at 375 K and 1 bar

```python
dataSolid=np.genfromtxt("Solid/histo")
dataLiquid=np.genfromtxt("Liquid/histo")
```

Plot distributions

```python
plt.fill_between(dataLiquid[:,0],0,dataLiquid[:,1],alpha=0.2)
plt.fill_between(dataSolid[:,0],0,dataSolid[:,1],alpha=0.2)
plt.plot(dataLiquid[:,0],dataLiquid[:,1])
plt.plot(dataSolid[:,0],dataSolid[:,1])

plt.xlim([0,1])
plt.ylim([0,5])
plt.xlabel(r"$\tilde{k}(\chi,\chi')$")
plt.ylabel("Probability")

plt.savefig("distributions.png", bbox_inches='tight')

plt.show()
```
